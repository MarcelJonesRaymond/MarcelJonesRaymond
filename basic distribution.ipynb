```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
x=stats.binom.pmf(n=10,p=0.6,k=np.arange(0,10))
```

```python
plt.plot(np.arange(0,10),x, marker="*")
plt.title(" Binormal Distribution")
```
