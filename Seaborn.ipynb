```python
s=sns.load_dataset("tips")
s.head(3)
```

```python
a=pd.crosstab(index=s["sex"],columns=s["smoker"],values=s["tip"],aggfunc="sum",margins=True)
a
```

```python
mapp=s.corr()
```

```python
s.cov()
```

```python
sns.heatmap(mapp,annot=True,cmap="coolwarm")
```

```python
pd.pivot_table(data=s,
    values="tip",
    index="sex",
    columns="day",
    aggfunc='sum',
    fill_value=None,
    margins=True)
```

# Pair Grid


```python
i=pd.read_csv("tips.csv")
```

```python
pairing=sns.PairGrid(i)
pairing.map_diag(sns.boxplot)
pairing.map_lower(sns.kdeplot)
pairing.map_upper(sns.scatterplot)
```

```python
h=sns.FacetGrid(data=i,
    row=None,
    col="sex",
    hue="smoker")
h.map(sns.distplot,"tip");
```

```python
q=sns.FacetGrid(data=i,row="size",
    col="time",
    hue="sex")
q.map(sns.scatterplot,"total_bill","tip")
```
